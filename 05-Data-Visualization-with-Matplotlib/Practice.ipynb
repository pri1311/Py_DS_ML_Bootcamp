```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
# Plots x vs y
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y)
axes1.set_title("Larger Plot")

axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes2.plot(y, x)
axes2.set_title("Smaller Plot")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
plt.tight_layout()
```

```python
x = np.random.rand(10)
y = np.random.rand(10)

plt.plot(x, y, 'g*-')
```

```python
fig = plt.figure(figsize=(8,2), dpi=100)
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(figsize=(8, 2), dpi=100, nrows = 2, ncols = 1)
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x, x ** 2, label = "x square")
axes.plot(x, x ** 3, label = "x cube")

axes.legend()
```

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 1, 1])

# axes.plot(x, y, color= "#1ca367") 
# accepts basic strings(green, purple, etc), rgb and hex codes.

axes.plot(x, y, color="#1ca367", linewidth = 2, alpha = 0.5, linestyle = "-", marker = "o", markersize = 10)

axes.set_xlim([0, 1])
axes.set_ylim([0, 2])
```
